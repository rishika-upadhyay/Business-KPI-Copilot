```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_excel("../data/Cleaned_Superstore.xlsx")
```

```python
df.head()
```

# Sales by Category 

```python
sales_category = df.groupby("Category")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
sales_category.plot(kind="bar", color=["#4E79A7","#59A14F","#F28E2B"])

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=0)

plt.savefig("../images/sales_by_category.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Profit by Category

```python
profit_category = df.groupby("Category")["Profit"].sum().sort_values()

plt.figure(figsize=(8,5))
profit_category.plot(kind="bar", color=["#E15759","#76B7B2","#59A14F"])

plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation=0)

plt.savefig("../images/profit_by_category.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Sales by Region

```python
sales_region = df.groupby("Region")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
sales_region.plot(kind="bar", color=["#E15759", "#F28E2B", "#76B7B2", "#4E79A7"])

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=0)

plt.savefig("../images/sales_by_region.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Profit by Region

```python
profit_region = df.groupby("Region")["Profit"].sum().sort_values()

plt.figure(figsize=(8,5))
profit_region.plot(kind="bar", color=["#FF6B6B", "#FFD166", "#06D6A0", "#118AB2"])

plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.xticks(rotation=0)

plt.savefig("../images/profit_by_region.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Top 10 States by Sales

```python
sales_state = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sales_state.plot(kind="bar", color="royalblue")

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")
plt.xticks(rotation=45)

plt.savefig("../images/top10_states_sales.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Top 10 States by Profit

```python
profit_state = df.groupby("State")["Profit"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
profit_state.plot(kind="bar", color="green")

plt.title("Top 10 States by Profit")
plt.xlabel("State")
plt.ylabel("Profit")
plt.xticks(rotation=45)

plt.savefig("../images/top10_states_profit.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Sales by Ship Mode

```python
sales_shipmode = df.groupby("Ship Mode")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
sales_shipmode.plot(kind="bar", color="orange")

plt.title("Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Sales")
plt.xticks(rotation=15)

plt.savefig("../images/sales_shipmode.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Profit by Ship Mode

```python
profit_shipmode = df.groupby("Ship Mode")["Profit"].sum().sort_values()

plt.figure(figsize=(8,5))
profit_shipmode.plot(kind="bar", color="purple")

plt.title("Profit by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Profit")
plt.xticks(rotation=15)

plt.savefig("../images/profit_shipmode.png", dpi=300, bbox_inches="tight")

plt.show()
```
