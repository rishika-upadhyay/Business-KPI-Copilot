```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel("../data/Cleaned_Superstore.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
conn = sqlite3.connect("../data/business_kpi.db")
```

```python
df.to_sql("sales_data", conn, if_exists="replace", index=False)
```

```python
query = "SELECT * FROM sales_data LIMIT 5;"
pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Sales) AS Total_Sales
FROM sales_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Profit) AS Total_Profit
FROM sales_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT ROUND(AVG(Sales),2) AS Average_Sales
FROM sales_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT ROUND(AVG(Profit),2) AS Average_Profit
FROM sales_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Quantity) AS Total_Quantity
FROM sales_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
       ROUND(SUM(Sales),2) AS Total_Sales
FROM sales_data
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```
