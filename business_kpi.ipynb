```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/Cleaned_Superstore.xlsx")
```

```python
df.head()
```

```python
df.shape
```

# Overall KPI Summary

```python
kpi_summary = pd.DataFrame({
    "KPI": [
        "Total Sales",
        "Total Profit",
        "Average Sales",
        "Average Profit",
        "Total Quantity Sold",
        "Average Discount"
    ],
    "Value": [
        round(df["Sales"].sum(), 2),
        round(df["Profit"].sum(), 2),
        round(df["Sales"].mean(), 2),
        round(df["Profit"].mean(), 2),
        int(df["Quantity"].sum()),
        round(df["Discount"].mean(), 2)
    ]
})

kpi_summary
```

```python
kpi_summary.to_excel("../data/kpi_summary.xlsx", index=False)
```

```python
import os

os.listdir("../data")
```

# Category KPI Summary

```python
category_kpi = df.groupby("Category").agg({
    "Sales":"sum",
    "Profit":"sum",
    "Quantity":"sum",
    "Discount":"mean"
}).reset_index()

category_kpi["Sales"] = category_kpi["Sales"].round(2)
category_kpi["Profit"] = category_kpi["Profit"].round(2)
category_kpi["Discount"] = category_kpi["Discount"].round(2)

category_kpi
```

```python
category_kpi.to_excel("../data/category_kpi_summary.xlsx", index=False)
```
