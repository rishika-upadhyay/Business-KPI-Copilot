```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_excel("../data/Cleaned_Superstore.xlsx")
```

```python
sales_category = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sales_category.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=0)

plt.savefig("../images/dashboard_sales_category.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
profit_category = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
profit_category.plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation=0)

plt.savefig("../images/dashboard_profit_category.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
sales_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sales_region.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=0)

plt.savefig("../images/dashboard_sales_region.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
profit_region = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
profit_region.plot(kind="bar")

plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.xticks(rotation=0)

plt.savefig("../images/dashboard_profit_region.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```
